```python
class data_retrival:
    def __init__(self) : 
        self.text_df = pd.DataFrame([])
        self.i = 1
        self.path = ""
        
    def load_df(self, data, category):
        for d in data:
            self.text_df = self.text_df.append(pd.DataFrame({'text': d['text'], 'category': category}, index=[0]), ignore_index=True)

    def get_data(self, path, category):
        data = []
        if os.path.exists(path):
            for line in open(path, 'r'):
                data.append(json.loads(line))
            self.load_df(data, category)
        else:
            print('No file by this name',path)
    def get_df(self):
        return self.text_df
```

# Final Preprocessing on text

```python
from io import StringIO
df = df[pd.notnull(df['text'])]
df['category_id'] = df['category'].factorize()[0]
category_id_df = df[['category', 'category_id']].drop_duplicates().sort_values('category_id')
category_to_id = dict(category_id_df.values)

id_to_category = dict(category_id_df[['category_id', 'category']].values)
df.head()
```

## Run either with regex or without regex. With regex takes more time

## With Regex  

```python

warnings.filterwarnings('ignore')
documents = []

from nltk.stem import WordNetLemmatizer

stemmer = WordNetLemmatizer()

for sen in range(0, df.shape[0]): 
    
    df.text[sen] = re.sub(r'\W', ' ', str(df.text[sen]))

    df.text[sen] = re.sub(r'\s+[a-zA-Z]\s+', ' ', df.text[sen])

    df.text[sen] = re.sub(r'\^[a-zA-Z]\s+', ' ', df.text[sen]) 

    df.text[sen] = re.sub(r'\s+', ' ', df.text[sen], flags=re.I)

    df.text[sen] = df.text[sen].lower()

    document = df.text[sen].split()

    document = [stemmer.lemmatize(word) for word in document]
    df.text[sen] = ' '.join(document)
```

# Validating Data (Takes Time on Local Machine (Run if you want more insights on current data) - Nothing to deal with classifier)

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf = TfidfVectorizer(sublinear_tf=True, min_df=5, norm='l2', encoding='latin-1', ngram_range=(1, 2), stop_words='english')
features = tfidf.fit_transform(df.text).toarray()
labels = df.category_id
features.shape
```

## Finding unigram and bigram correlation

```python
from sklearn.feature_selection import chi2

N = 2
for category, category_id in sorted(category_to_id.items()):
    features_chi2 = chi2(features, labels == category_id)
    indices = np.argsort(features_chi2[0])
    feature_names = np.array(tfidf.get_feature_names())[indices]
    unigrams = [v for v in feature_names if len(v.split(' ')) == 1]
    bigrams = [v for v in feature_names if len(v.split(' ')) == 2]
    print("# '{}':".format(category))
    print("  . Most correlated unigrams:\n. {}".format('\n. '.join(unigrams[-N:])))
    print("  . Most correlated bigrams:\n. {}".format('\n. '.join(bigrams[-N:])))
    
```

# Modeling - Multinomial Naive Bayes

```python
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfTransformer
from sklearn.naive_bayes import MultinomialNB

X_train, X_test, y_train, y_test = train_test_split(df['text'], df['category'], random_state = 0)
count_vect = CountVectorizer()
X_train_counts = count_vect.fit_transform(X_train)
tfidf_transformer = TfidfTransformer()
X_train_tfidf = tfidf_transformer.fit_transform(X_train_counts)

clf = MultinomialNB().fit(X_train_tfidf, y_train)
```
